```python
import requests
import pandas as pd
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://swapi.dev/api/people/'
response = requests.get(url)
json = response.json()
json
```

```python
json.keys()
```

```python
people = pd.DataFrame(json['results'])
people
```

```python
while json['next'] != None:
    print(json['next'])
    response = requests.get(json['next'])
    json = response.json()
    people = pd.concat([people, pd.DataFrame(json['results'])], ignore_index = True)
```

```python
url = 'https://swapi.dev/api/planets/'
response = requests.get(url)
json = response.json()
json.keys()
```

```python
planets = pd.DataFrame(json['results'])
planets
```

```python
url = 'https://swapi.dev/api/starships/'
response = requests.get(url)
json = response.json()
json.keys()
```

```python
starships = pd.DataFrame(json['results'])
starships
```

```python
people.to_csv('people.csv', index = False)
planets.to_csv('planets.csv', index = False)
starships.to_csv('starships.csv', index = False)
```

```python
pd.options.display.max_columns = None
starwars_df = pd.concat([people, planets, starships], axis = 1)
starwars_df
```

```python
def get_people_data():
    
    if os.path.isfile('people.csv'):
        
        return pd.read_csv('people.csv')
    
    else:
       
        url = 'https://swapi.dev/api/people/'
        response = requests.get(url)
        json = response.json()

        people = pd.DataFrame(json['results'])
        
        people.to_csv('people.csv', index = False)
   
    return people
```

```python
get_people_data()
```

```python
def get_planets_data():
    
    if os.path.isfile('planets.csv'):
        
        return pd.read_csv('planets.csv')
    
    else:
       
        url = 'https://swapi.dev/api/planets/'
        response = requests.get(url)
        json = response.json()

        planets = pd.DataFrame(json['results'])
        
        planets.to_csv('planets.csv', index = False)
   
    return planets
```

```python
get_planets_data()
```

```python
def get_starships_data():
    
    if os.path.isfile('starships.csv'):
        
        return pd.read_csv('starships.csv')
    
    else:
       
        url = 'https://swapi.dev/api/starships/'
        response = requests.get(url)
        json = response.json()

        starships = pd.DataFrame(json['results'])
        
        starships.to_csv('people.csv', index = False)
   
    return starships
```

```python
get_starships_data()
```

```python
def combine_df(df_x, df_y, df_z):
    
    pd.options.display.max_columns = None
    
    starwars_df = pd.concat([df_x, df_y, df_z], axis = 1)
    
    return starwars_df
```

```python
combine_df(people, starships, planets)
```

```python
def get_renewable_energy_data():
    
    if os.path.isfile('renewable_energy_production.csv'):
        
        return pd.read_csv('renewable_energy_production.csv')
    
    else:
       
        url = 'https://raw.githubusercontent.com/jenfly/opsd/master/opsd_germany_daily.csv'

        df = pd.DataFrame(pd.read_csv(url))
        
        df.to_csv('renewable_energy_production.csv', index = False)
   
    return df
```
